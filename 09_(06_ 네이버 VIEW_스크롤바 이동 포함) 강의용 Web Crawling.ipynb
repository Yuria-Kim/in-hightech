<span style="color: blue; font-size: 12pt">
● driver.execute_script 함수에서 절대 좌표를 기준으로 스크롤바를 내리는 window.scrollTo() 사용법 <br>

</span>

사용법 예시   
#x축 좌표는 0, y축 좌표는 500픽셀 위치까지 스크롤   
driver.execute_script("window.scrollTo(0, 500)") 

 #x축 좌표는 0, y축 좌표는 웹 페이지 맨 아래까지 스크롤  
driver.execute_script("window.scrollTo(0, document.body.scrollHeight)")

<span style="color: blue; font-size: 12pt">
● driver.execute_script 함수에서 상대 좌표를 기준으로 스크롤바를 내리는 window.scrollBy() 사용법 <br> 
driver.execute_script("window.scrollBy( x 좌표 , y 좌표) " )  
</span>

사용법 예시   
#현재 위치에서 x축 좌표는 스크롤하지 않고, y축으로만 500픽셀 아래로 스크롤   
driver.execute_script("window.scrollBy(0, 500)") 

```python
# 함수 정의
def scroll_down(driver):
    driver.execute_script("window.scrollTo(0,document.body.scrollHeight)")
    time.sleep(3)      
```

```python
real_cnt = 6  #스크롤바 내리기 회수 4회 셋팅  
i = 1

while (i <= real_cnt) :  #1~6회 스크롤바 내리기 
    scroll_down(driver)
    i += 1    # i = i+1 과 동일    
```

<span style="color: red; font-size: 18pt">
Step 11. 출력 결과를 리스트로 저장한 후 txt 파일 형식의 데이터셋 생성하기
</span> 

```python
# 저장할 리스트 선언 

no_list=[]           # 게시글 번호 컬럼
title_list=[ ]       # 게시글 제목 컬럼
contents_list=[ ]    # 게시글 내용 컬럼
name_list=[ ]        # 작성자 컬럼
date_list=[ ]        # 작성일 컬럼

#스크롤바를 내려서 출력되고 있는 전체 내용을 크롤링 
html = driver.page_source
soup = BeautifulSoup(html, 'html.parser')

blog_list = soup.find_all('div', 'total_wrap api_ani_send')  
print("● blog_list " )
print( blog_list )
print("■"*50)

no = 1        # 게시글 번호 초기값 
f = open(ft_name, 'w',encoding='UTF-8')  #txt 파일 오픈 

for i in blog_list: 
    if no <= cnt : 

        # 1.게시글 번호        
        no_list.append(no)   #no_list 리스트에 카운트 번호 추가  
        f.write("번호 :" + str(no) + "\n") #txt 파일에 카운트 번호 기록 
        print("번호 : %s" %no)

        # 2. 게시글 제목
        title=i.find('a','api_txt_lines total_tit').get_text()
        title = title.strip() 
        title_list.append(title)  #title_list 리스트에 제목 추가  
        f.write("제목 :" + str(title) + "\n") #txt 파일에 제목 기록 
        print("제목 : %s" %title)

        # 3. 게시글 내용
        contents = i.find('div','api_txt_lines dsc_txt').get_text().replace("\n","")   
        contents = contents.strip() 
        contents_list.append(contents)   #contents_list 리스트에 내용 추가  
        f.write("내용 :" + str(contents) + "\n") #txt 파일에 내용 기록 
        print("내용 : %s" %contents)

        # 4. 게시글 작성자
        blog_name = i.find('a','sub_txt sub_name').get_text()
        name_list.append(blog_name)   #name_list 리스트에 작성자 추가  
        f.write("작성자 :" + str(blog_name) + "\n")  #txt 파일에 작성자 기록 
        print("작성자 : %s" %blog_name)

        # 5. 게시글 작성일
        blog_date = i.find('span','sub_time sub_txt').get_text()
        date_list.append(blog_date)   #date_list 리스트에 작성일 추가  
        f.write("작성일 :" + str(blog_date) + "\n\n")     #txt 파일에 작성일 기록     
        print("작성일 : %s \n" %blog_date)
        
        no += 1

f.close( )
```

```python
print("크롤링할 데이터 개수 : ",  cnt)
print("최종 크롤링된 데이터 개수 : ",  no-1)
```

Step 12. 출력 결과를 pandas를 이용하여 데이터프레임으로 생성하기

```python
#위에서 import pandas as pd  선언했기 때문에 pd.DataFrame()를 사용할 수 있다. 

naver_blog = pd.DataFrame()

naver_blog['번호']=no_list
naver_blog['제목']=title_list
naver_blog['내용']=contents_list
naver_blog['작성자']=name_list
naver_blog['작성일']=date_list
```

```python
#csv 형태로 저장하기
naver_blog.to_csv(fc_name, encoding="UTF-8-SIG", index=False)

#엑셀 형태로 저장하기
naver_blog.to_excel(fx_name, encoding="UTF-8-SIG", index=False)

e_time = time.time( )     # 검색이 종료된 시점의 timestamp 를 지정합니다
t_time = e_time - s_time 
```

● 정적 페이지란?   
정적페이지란 항상 같은 내용을 보여주는 웹페이지로 사용자가 URL을 통해 서버에 웹 페이지를 요청하였을 때,   
서버 안에 이미 만들어져 있는 HTML 문서를 사용자에게 보여주는 경우 이를 정적 웹페이지라고 한다.  
ex) 백과사전의 내용과 같이 항상 같은 내용을 보여준다.  

● 동적 페이지란?  
동적페이지란 요청에 대해서 각각 다른 내용을 보여준다.  
사용자(클라이언트)가 URL 을 통해 서버에 웹 페이지를 요청했을 때,   
서버는 사용자에 맞는 HTML 문서를 생성하여 사용자에게 응답하게 된다. 이를 동적 웹 페이지라고 한다.  
ex) 온라인 쇼핑몰의 장바구니, 최근 본 상품, 추천 물건, 사이트의 회원가입, 로그인 등
